## $\S$ Quaternion

### Def. 1

For $p, q \in \RR^4$, define:

$~~
\bba
p \odot q
    &\equiv
        \pmatrix{
            p_0 q_0 - \vec{p}_{123} \cdot \vec{q}_{123} \\ 
            p_0 \vec{q}_{123} + q_0 \vec{p}_{123} +
                \vec{p}_{123} \xx \vec{q}_{123}
        } \\ \\
    &=
        \pmatrix{
            p_0,           & - \vec{p}_{123}^T  \\ 
            \vec{p}_{123}, & p_0 I + C_{\vec{p}_{123}}
        } q
    \; \equiv L_p q \\ \\
    &=
        \pmatrix{
            q_0,           & - \vec{q}_{123}^T  \\ 
            \vec{q}_{123}, & q_0 I + C_{- \vec{q}_{123}}
        } p
    \; \equiv R_q p,
\eea
$

where

$~~
C_{u} = \pmatrix{0 & u_z & - u_y \\ -u_z & 0 & u_x \\ u_y & -u_x & 0}.
$

### Def. 2

Define $(-)^\dagger$ by:

$~~
p^{\dagger}
    \equiv \pmatrix{ p_0 \\ - \vec{p}_{123} }
    = \pmatrix{1, & \vec{0}^T \\ \vec{0}, & - I_3 } p
    \equiv N p.
$


### Prop. 3

$~~ (p \odot q)^\dagger = q^\dagger \odot p^\dagger$.

***Proof***

We see:

$~~
\bba
N R_q N
    &= 
        N
        \pmatrix{
            q_0,           & - \vec{q}_{123}^T  \\ 
            \vec{q}_{123}, & q_0 I + C_{- \vec{q}_{123}}
        }
        N
    =
        \pmatrix{
            q_0,           & \vec{q}_{123}^T  \\ 
            - \vec{q}_{123}, & q_0 I + C_{- \vec{q}_{123}}
        }
    = L_{q^\dagger} \\ \\
(p \odot q)^\dagger
    &= N R_q p = (N R_q N) \; (N p) = L_{q^\dagger} p^\dagger
    = q^\dagger \odot p^\dagger.
\eea
$

### Prop. 4

$~~
p \odot (q \odot r) = (p \odot q) \odot r
$

***Proof***

This amounts to show $L_p L_q = L_{p \odot q}$.

We see:

$
\bba
L_p L_q
    &=
        \pmatrix{
            p_0,           & - \vec{p}_{123}^T  \\ 
            \vec{p}_{123}, & p_0 I + C_{\vec{p}_{123}}
        }
        \pmatrix{
            q_0,           & - \vec{q}_{123}^T  \\ 
            \vec{q}_{123}, & q_0 I + C_{\vec{q}_{123}}
        } \\ \\
    &=
        \pmatrix{
            p_0 q_0 - \vec{p}_{123} \cdot \vec{q}_{123}, &
                - p_0 \vec{q}_{123}^T - q_0 \vec{p}_{123}^T
                    - \vec{p}_{123}^T C_{\vec{q}_{123}}  \\
            q_0 \vec{p}_{123} + p_0 \vec{q}_{123}
                + C_{\vec{p}_{123}} \vec{q}_{123},  &
                    - \vec{p}_{123} \vec{q}_{123}^T
                    + p_0 q_0 I
                    + p_0 C_{\vec{q}_{123}}
                    + q_0 C_{\vec{p}_{123}}
                    + C_{\vec{p}_{123}} C_{\vec{q}_{123}}                    
        }.
\eea
$

Also noting that:

$~~
\bba
C_u C_v x
    &= u \xx (v \xx x)
    = (u \cdot x) v - (u \cdot v) x
    = \ll( v \, u^T - v \cdot u \, I \rr) \, x
\\
C_{u \xx v} x
    &= (u \xx v) \xx x
    = - x \xx (u \xx v)
    = - (x \cdot v) u + (x \cdot u) v
    = \ll( - u v^T + v u^T \rr) x
\\
v^T C_u &= \ll( C_u^T v \rr)^T
    = - \ll( C_u v \rr)^T = - (u \xx v)^T = (v \xx u)^T
    = \ll( C_v u \rr)^T
\eea
$

we have:

$~~
\bba
- \vec{p}_{123} &\vec{q}_{123}^T
+ p_0 q_0 I
+ p_0 C_{\vec{q}_{123}}
+ q_0 C_{\vec{p}_{123}}
+ C_{\vec{p}_{123}} C_{\vec{q}_{123}}  \\
    &=
        - \vec{p}_{123} \vec{q}_{123}^T
        + p_0 q_0 I
        + p_0 C_{\vec{q}_{123}}
        + q_0 C_{\vec{p}_{123}}
        + \vec{q}_{123} \vec{p}_{123}^T
        - (\vec{p}_{123} \cdot \vec{q}_{123}) I  \\
    &=
        (p_0 q_0 - \vec{p}_{123} \cdot \vec{q}_{123}) I
        + p_0 C_{\vec{q}_{123}}
        + q_0 C_{\vec{p}_{123}}
        - \vec{p}_{123} \vec{q}_{123}^T
        + \vec{q}_{123} \vec{p}_{123}^T  \\
    &=
        (p_0 q_0 - \vec{p}_{123} \cdot \vec{q}_{123}) I
        + C_{p_0 \vec{q}_{123}}
        + C_{q_0 \vec{p}_{123}}
        + C_{\vec{p}_{123} \xx \vec{q}_{123}}.
\eea
$

Thus:

$~~
\bba
L_p L_q
    &=
        \pmatrix{
            p_0 q_0 - \vec{p}_{123} \cdot \vec{q}_{123}, &
                - p_0 \vec{q}_{123}^T - q_0 \vec{p}_{123}^T
                    - \ll( C_{\vec{p}_{123}} \vec{q}_{123} \rr)^T \\
            q_0 \vec{p}_{123} + p_0 \vec{q}_{123}
                + C_{\vec{p}_{123}} \vec{q}_{123},  &
                    \ll( p_0 q_0 - \vec{p}_{123} \cdot \vec{q}_{123} \rr) I
                    + C_{
                        p_0 \vec{q}_{123}
                        + q_0 \vec{p}_{123}
                        + \vec{p}_{123} \xx \vec{q}_{123}}
        } \\ \\
    &= L_{p \odot q}.
\eea
$

### Def. 5

Given $p \in S^3$, define $A_p : \RR^3 \to \RR^3$ by:

$~~ A_p(v) = \ll(p \odot \pmatrix{0 \\ v} \odot p^{\dagger}\rr)_{123}$.

Here note that $\ll( A_p(v) \rr)_0 = 0$ since

$
p \odot \pmatrix{0 \\ v} \odot p^{\dagger}
+ \ll( p \odot \pmatrix{0 \\ v} \odot p^{\dagger} \rr)^\dagger
    =
        p \odot \pmatrix{0 \\ v} \odot p^{\dagger}
        + p \odot \pmatrix{0 \\ - v} \odot p^{\dagger}
    = 0.
$

### Prop. 6

$~~
A_p = u u^T + \cos(\phi) \, (- C_u C_u) + \sin(\phi) C_u
    \in SO(3),
$

where

$~~
p = \pmatrix{
        \cos\ll(\dfrac{\phi}{2}\rr) \\
        \sin\ll(\dfrac{\phi}{2}\rr) \; u
    }
\quad \tt{and} ~~~ |u| = 1.
$

***Proof***

First, we show $A_p \in SO(3)$.

Noting that:

$~~
u^T C_u x = u \cdot (u \xx x) = 0 \\
y^T C_u^T C_u C_u x = (u \xx y) \cdot (u \xx (u \xx x)) = 0,
$

we have:

$~~
\bba
A_p^T A_p
    &= u u^T u u^T
        + \cos^2(\phi) \; (C_u C_u)^T C_u C_u
        + \sin^2(\phi) \; C_u^T C_u  \\
    &= u u^T + \cos^2(\phi) C_u^4 - \sin^2(\phi) C_u^2 \\
    &= u u^T + \cos^2(\phi) (u u^T - I)^2 - \sin^2(\phi) (u u^T - I) \\
    &= u u^T + \cos^2(\phi) (I - u u^T) - \sin^2(\phi) (u u^T - I) \\
    &= I.
\eea
$

Next, we will show $A_p = u u^T + \cos(\phi) \, (- C_u C_u) + \sin(\phi) C_u$.

First, we note:

$~~
p \odot q \odot p^\dagger
    = (p \odot (p \odot q)^\dagger)^\dagger
    = N L_p N L_p q = (N L_p)^2 q.
$

Then, here we have:

$~~
\bba
(N L_p)^2
    &= \pmatrix{
        p_0,             & - \vec{p}_{123}^T  \\ 
        - \vec{p}_{123}, & - p_0 I - C_{\vec{p}_{123}}
        }^2 \\
    &= \pmatrix{
        *, & *  \\ 
        *, &
            \vec{p}_{123} \vec{p}_{123}^T
            + p_0^2 I
            + 2 p_0 C_{\vec{p}_{123}}
            + C_{\vec{p}_{123}}^2
        }
      \quad (\tt{where "$*$" : we don't care}).
\eea
$

and

$~~
\bba
A_p &=
        \vec{p}_{123} \vec{p}_{123}^T
        + p_0^2 I
        + 2 p_0 C_{\vec{p}_{123}}
        + C_{\vec{p}_{123}}^2 \\
    &=
        \sin^2(\phi/2) u u^T
        + \cos^2(\phi/2) \; I
        + 2 \sin(\phi/2) \cos(\phi/2) \; C_u
        + \sin^2(\phi/2) \; C_u^2 \\
    &=
        u u^T
        - \cos^2(\phi/2) u u^T
        + \cos^2(\phi/2) I
        + \sin(\phi) C_u
        + \sin^2(\phi/2) \; (u u^T - I) \\
    &=
        u u^T
        + \sin(\phi) C_u
        + \ll( \sin^2(\phi/2) - \cos^2(\phi/2) \rr) (u u^T - I) \\        
    &=
        u u^T
        + \sin(\phi) C_u
        - \cos(\phi) C_u C_u.
\eea
$

$~~ A_p = A_q \implies p = q ~~ \tt{or} ~~ p = - q$.

Note that its converse is trivial.

***Proof***

We write $p = \pmatrix{s \\ v}$, then

$~~
\bba
A_p &= v v^T + s^2 I + 2 s C_v + C_v^2  \\

$~~

## $\S$ Uniform distribution on $S^3$

### Prop. 1


Define $F : [0, 1) \xx [0, 2\pi) \xx [0, 2 \pi) \to S^3$ by:

$~~
F(u, \phi_1, \phi_2) \equiv
    \pmatrix{
        u^{\frac{1}{2}} \cos(\phi_1) \\
        u^{\frac{1}{2}} \sin(\phi_1) \\
        (1 - u)^{\frac{1}{2}} \cos(\phi_2) \\
        (1 - u)^{\frac{1}{2}} \sin(\phi_2) \\
    } =
    \pmatrix{
        \sqrt{u} \;
            \pmatrix{ \cos(\phi_1) \\ \sin(\phi_1) } \\
        \sqrt{1 - u}
            \pmatrix{ \cos(\phi_2) \\ \sin(\phi_2) }
    }.    
$

Then, for the Jacobian $J = DF$, we have $\sqrt{|J^T J|} = \frac{1}{2}$ : constant.

***Proof***

We compute each partial derivative:

$~~
\bba
\pp_u F &= 
    \pmatrix{
        \pp_u \sqrt{u} \;
            \pmatrix{ \cos(\phi_1) \\ \sin(\phi_1) } \\
        \pp_u \sqrt{1 - u} \;            
            \pmatrix{ \cos(\phi_2) \\ \sin(\phi_2) }
    }
    =
        \pmatrix{
            \dfrac{1}{2 \sqrt{u}} \;
                \pmatrix{ \cos(\phi_1) \\ \sin(\phi_1) } \\
            \dfrac{-1}{2 \sqrt{1 - u}} \;            
                \pmatrix{ \cos(\phi_2) \\ \sin(\phi_2) }
        }    
    \\ \\
\pp_{\phi_1} F &= 
    \pmatrix{
        \sqrt{u} \;
            \pmatrix{ - \sin(\phi_1) \\ \cos(\phi_1) } \\
        0 \\ 0
    } \\ \\
\pp_{\phi_2} F &= 
    \pmatrix{
        0 \\ 0 \\
        \sqrt{1 - u} \;            
            \pmatrix{ - \sin(\phi_2) \\ \cos(\phi_2) }
    }.
\eea
$

Here, clearly
$~
\pp_u F \cdot \pp_{\phi_1} F
    = \pp_{\phi_1} F \cdot \pp_{\phi_2} F
    = \pp_{\phi_2} F \cdot \pp_{u} F
    = 0.
$

Therefore,

$~~
\bba
J^T J
    &=
        \pmatrix{
            |\pp_u F|^2 & 0                  & 0  \\
            0           & |\pp_{\phi_1} F|^2 & 0  \\
            0           & 0                  & |\pp_{\phi_2} F|^2  \\
        } \\ \\
    &=
        \pmatrix{
            (\pp_u \sqrt{u})^2 + (\pp_u \sqrt{1 - u})^2 & 0 & 0  \\
            0 & u & 0  \\
            0 & 0 & 1 - u  \\
        } \\ \\
    &=
        \pmatrix{
            \dfrac{1}{4 u} + \dfrac{1}{4 (1 - u)}  & 0 & 0  \\
            0 & u & 0  \\
            0 & 0 & 1 - u  \\
        } \\ \\
    &=
        \pmatrix{
            \dfrac{1}{4 u (1 - u)} & 0 & 0  \\
            0 & u & 0  \\
            0 & 0 & 1 - u  \\
        } \\
|J^T J| &= \frac{1}{4 u (1 - u)} u (1 - u) = \frac{1}{4}.
\eea
$

## $\S$ Stereographic projection


### Def. 1

We define $F$ as below:

$~~
\bba
F : \RR^{n + 1} \supset
        S^n \setminus \{ (1, \vec{0}) \}  &\to     \RR^n  \\
    (x_0, \vec{x}) \quad\quad             &\mapsto \vec{y} = \frac{\vec{x}}{1 - x_0}.
\eea
$


### Prop. 2

The inverse of $F$ is:

$~~
\bba
F^{-1}(\vec{y}) =
    \ll(
        \frac{|\vec{y}|^2 - 1}{|\vec{y}|^2 + 1},
        \frac{2 \vec{y}}{|\vec{y}|^2 + 1}
    \rr)
    \equiv G(\vec{y}).
\eea
$

***Proof***

Given $x_0^2 + |\vec{x}|^2 = 1$, we see:

$~~
\bba
\vec{y} = F(x_0, \vec{x}) = \frac{\vec{x}}{1 - x_0}
    &\implies
        |\vec{y}|^2
            = \frac{|\vec{x}|^2}{(1 - x_0)^2}
            = \frac{1 - x_0^2}{(1 - x_0)^2}
            = \frac{1 + x_0}{1 - x_0} \\
    &\implies
        (1 - x_0) |\vec{y}|^2 = 1 + x_0 \\
    &\implies
        x_0 = \frac{|\vec{y}|^2 - 1}{|\vec{y}|^2 + 1}
            = 1 - \frac{2}{|\vec{y}|^2 + 1}.
\eea
$

And, we also have:

$~~
\vec{x} = (1 - x_0) \vec{y} = \dfrac{2}{|\vec{y}|^2 + 1} \vec{y}
$.

So, we have proved $F^{-1} \circ F = \tt{id}$.

For the converse $F \circ F^{-1} = \tt{id}$, we have:

$~~
\bba
\frac{\vec{x}}{1 - x_0}
    =
        \ddfrac
            { \frac{2 \vec{y}}{|\vec{y}|^2 + 1} }
            { 1 - \frac{|\vec{y}|^2 - 1}{|\vec{y}|^2 + 1} }
    = 
        \ddfrac
            {  2 \vec{y} }
            { |\vec{y}|^2 + 1 - (|\vec{y}|^2 - 1)  }
    = \vec{y}.
\eea
$

### Prop. 3

For the Jacobian $J \equiv DG|_{\vec{y}} \in \RR^{(n + 1) \times n}$,
we have $\sqrt{|J^{T} J|} = \dfrac{2}{|\vec{y}|^2 + 1}$ (aka. pull-back volume form).

***Proof***

First we note:

$~~
\pp_\vec{y} \dfrac{1}{|\vec{y}|^2 + 1}
    =  - 2 \dfrac{\vec{y}^T}{(|\vec{y}|^2 + 1)^2} \in \RR^{1 \times n}.
$

Thus, writing 
$
DG|_{\vec{y}}
    = \pmatrix{
        \pp_\vec{y} x_0     &\in \RR^{1 \times n} \\
        \pp_\vec{y} \vec{x} &\in \RR^{n \times n}
      },
$
we see:

$~~
\bba
\pp_\vec{y} x_0
    &= \pp_\vec{y} \ll( 1 - \frac{2}{|\vec{y}|^2 + 1} \rr)
    = 4 \dfrac{\vec{y}^T}{(|\vec{y}|^2 + 1)^2} \\
\
\pp_\vec{y} \vec{x}
    &= \pp_\vec{y} \frac{2 \vec{y}}{|\vec{y}|^2 + 1}
    = 2 \ll( 
            \frac{1}{|\vec{y}|^2 + 1} I + 
            \vec{y} \; \pp_\vec{y} \frac{1}{|\vec{y}|^2 + 1}
        \rr) \\
    &= 2 \ll( 
            \frac{1}{|\vec{y}|^2 + 1} I
            - 2 \frac{\vec{y} \vec{y}^T}{(|\vec{y}|^2 + 1)^2}
        \rr).
\eea
$

Therefore:

$~~
\bba
J^T J &=
    \pmatrix{
        (\pp_\vec{y} x_0)^T \;
        (\pp_\vec{y} \vec{x})^T
      }
    \pmatrix{
        \pp_\vec{y} x_0 \\
        \pp_\vec{y} \vec{x}
      }
    = 
        (\pp_\vec{y} x_0)^T \pp_\vec{y} x_0
        + (\pp_\vec{y} \vec{x})^T \pp_\vec{y} \vec{x}  \\
    &=
        4^2 \dfrac{\vec{y} \vec{y}^T}{(|\vec{y}|^2 + 1)^4}
        + 2^2
            \ll( 
                \frac{1}{|\vec{y}|^2 + 1} I
                - 2 \frac{\vec{y} \vec{y}^T}{(|\vec{y}|^2 + 1)^2}
            \rr)^2 \\
    &=
        4 \frac{1}{(|\vec{y}|^2 + 1)^2}
            \ll(
                4 \dfrac{\vec{y} \vec{y}^T}{(|\vec{y}|^2 + 1)^2} +
                \ll( 
                    I - 2 \frac{\vec{y} \vec{y}^T}{|\vec{y}|^2 + 1}
                \rr)^2
            \rr)  \\
    &=
        4 \frac{1}{(|\vec{y}|^2 + 1)^2}       
            \ll(
                4 \dfrac{\vec{y} \vec{y}^T}{(|\vec{y}|^2 + 1)^2} +
                I + 4 |\vec{y}|^2 \frac{\vec{y} \vec{y}^T}{(|\vec{y}|^2 + 1)^2}
                - 4 \frac{\vec{y} \vec{y}^T}{|\vec{y}|^2 + 1}
            \rr)  \\            
    &=
        4 \frac{1}{(|\vec{y}|^2 + 1)^2}       
            \ll(
                I +
                4 \vec{y} \vec{y}^T
                    \ll(
                        \dfrac{1}{(|\vec{y}|^2 + 1)^2} +
                        |\vec{y}|^2 \frac{1}{(|\vec{y}|^2 + 1)^2}
                        - \frac{1}{|\vec{y}|^2 + 1}                    
                    \rr)
            \rr) \\
    &= \frac{4}{(|\vec{y}|^2 + 1)^2} I
    =  \ll( \frac{2}{\vec{y}|^2 + 1} \rr)^2 I.
\eea
$
