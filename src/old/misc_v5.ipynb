## van der Corput sequence

Define $\text{Corput} : \mathbb{N} \to [0, 1)$ by:

- $\text{Corput}(n) \equiv \sum_i \frac{n_i}{2^{i + 1}} = \text{"}0 . n_0 ... n_{k-1}\!\text{"}$,

where

- $n = \text{"}n_{k-1} ... n_{0}\!\text{"} = \sum_i n_i 2^{i} \le 2^k$.


### Prop. 1.

Such restriction as below is bijection:

- $\text{Corput} : \{0, 1, ..., 2^k - 1\} \cong \{\frac{0}{2^k}, \frac{1}{2^k}, .., \frac{2^k - 1}{2^k}\}$.

_Proof_

Due to bit reversal, this is obvious.

### Prop. 2.

Given

- $n \in \mathbb{N}$,
- $m \lt 2^k$,
- $p = \text{Corput}^{-1}(\frac{m}{2^k}) \lt 2^k$,


then, 

$
\newcommand{\Corput}[0]{\text{Corput}}
\begin{align}
n \equiv p \mod 2^k
    &\Longleftrightarrow
        \Corput(n) \in \left[ \frac{m}{2^k}, \frac{m + 1}{2^k} \right)
\\
    &\Longleftrightarrow
        \Corput(n) - \frac{m}{2^k} \in \left[ 0, \frac{1}{2^k} \right).
\end{align}
$

_Proof_

We write:

- $n = n_{l} ... n_{k} n_{k-1} ... n_0$,

then,

- $\Corput(n) = 0.n_0...n_{k-1} n_{k} ... n_l = 0.n_0...n_{k-1} + \frac{0.n_{k} ... n_l}{2^{k}}$.

Here noting,

- $\Corput(n_{k-1} ... n_0) = 0.n_0...n_{k-1} = \frac{m'}{2^k}$ for some $m'$,
- $\frac{0.n_{k} ... n_l}{2^{k}} \lt \frac{1}{2^k}$,

we have:

$
\begin{align}
\Corput(n) \in \left[ \frac{m}{2^k}, \frac{m + 1}{2^k} \right)
    &\;\Leftrightarrow\;
         \frac{m}{2^k} - \frac{0.n_{k} ... n_l}{2^{k}}
             \le
                 \frac{m'}{2^k}
                     \lt
                         \frac{m}{2^k} + \frac{1 - 0.n_{k} ... n_l}{2^{k}}
\\
    &\;\Leftrightarrow\;
         \Corput(n_{k-1} ... n_0)  = \frac{m'}{2^k} = \frac{m}{2^k} = \Corput(p)
\\
    &\;\Leftrightarrow\;
        n_{k-1} ... n_0 = p
\\
    &\;\Leftrightarrow\;
        n = p \mod 2^k
\end{align}
$.

### Def. 3.

For $S \subseteq [0, 1]$, we define:

- $A_N(S) \equiv \# \{ n \lt N \mid \text{Corput}(n) \in S \}$.

As a special case, for $x \in [0, 1)$, we define:

- $B_N(x) = A_N([0, x))$.


### Prop. 4.

$\# \{ n \lt N \mid n = p \mod L\} = ⌊ \frac{N - p - 1}{L} ⌋ + 1$.

_Proof_ : it can be seen easily.



### Prop. 5.

For $x \in [0, 1)$,

- $\lim_{N \to \infty} \frac{B_N(x)}{N} = x$ (aka. "uniform distribution modulo 1")


_Proof_ :

For $m \lt 2^k$, we take $p = \Corput^{-1}(\frac{m}{2^k})$.
Then, from above propositions, we have:

- $
A_N([\frac{m}{2^k}, \frac{m+1}{2^k}))
    = ⌊ \frac{N - p - 1}{2^k} ⌋ + 1
    = \frac{N - p - 1}{2^k} - \{ \frac{N - p - 1}{2^k} \} + 1
    = \frac{N}{2^k} + O(1).
$

Thus, 

- $
\begin{align}
\lim_{N \to \infty} \frac{A_N([\frac{m}{2^k}, \frac{m+1}{2^k}))}{N}
    &=
        \frac{1}{2^k}.
\end{align}
$

And therfore, 

- $
\lim_{N \to \infty}
    B_N(\frac{m}{2^k})
        = \sum_{i \lt m}
            \lim_{N \to \infty}
                A_N([\frac{i}{2^k}, \frac{i+1}{2^k}))
        = \frac{m}{2^k}
$.

Now, since $\{\frac{m}{2^k}\}_{m, k}$ is dense in $[0, 1)$,
we can find $q_i, r_i \in \{\frac{m}{2^k}\}_{m, k}$ s.t. 

- $\lim_i q_i = \lim_i r_i = x$, and
- $q_i \le x \lt r_i$.

So, our result limit follows from sandwitching via $B_N(q_i) \le B_N(x) \le B_N(r_i)$.

### Prop. 6.

For $x \in [0, 1)$,

- $|B_N(x) - N x| \le \log_2(N) + 2$.

_Proof_

We take $k$ s.t. $2^{k-1} \lt N \le 2^k$.

We write $x = 0.a_0...a_{k-1} a_{k} ...$.

Then, for $n = n_{k-1}...n_0 \lt N \le 2^k$, we have:

$
\begin{align}
\Corput(n) \lt x
    &\;\Longleftrightarrow\;
        0.n_0...n_{k-1} \lt 0.a_0...a_{k-1}
\\
    &\;\Longleftrightarrow\;
        (a_0 = 1 \wedge n_0 = 0) \;\vee\; (a_0 = n_0 \wedge 0.n_1...n_{k-1} \lt 0.a_1...a_{k-1})
\\
    &\;\Longleftrightarrow\;
        \begin{cases}
            (a_0 = 1 \wedge n_0 = 0) \;\vee\;
        \\
            (a_0 = n_0 \wedge a_1 = 1 \wedge n_1 = 0) \;\vee\;
        \\
            (a_0 = n_0 \wedge a_1 = n_1 \wedge 0.n_2...n_{k-1} \lt 0.a_2...a_{k-1}) \;\vee\;
        \end{cases}
\\
    &\;\Longleftrightarrow\;
        \begin{cases}
            (a_0 = 1 \wedge n_0 = 0) \;\vee\;
        \\
            (a_1 = 1 \wedge n_1 n_0 = a_0) \;\vee\;
        \\
            ...
        \\
            (a_{k-1} = 1 \wedge n_{k-1}...n_1 n_0 = a_{k-2}...a_0)
        \end{cases}
\\
    &\;\Longleftrightarrow\;
        \begin{cases}
            (a_0 = 1 \;\wedge\; n = 0 \!\mod 2) \;\vee\;
        \\
            (a_1 = 1 \;\wedge\; n = a_0 \!\mod 2^2) \;\vee\;
        \\
            ...
        \\
            (a_{k-1} = 1 \;\wedge\; n = a_{k-2}...a_0 \!\mod 2^{k-1}).
        \end{cases}
\end{align}
$

Importantly, notice that "$\vee$ predicates" in the above equivalence is exclusive, thus we have:

$
\begin{align}
B_N(x)
    &=
        \sum_{i = 0}^{k-1} a_i \cdot \# \{n \lt N \mid n = a_{i-1}..a_0 \!\mod 2^i \}
\\
    &=
        \sum_{i = 0}^{k-1} a_i \cdot
            \left(⌊\frac{N -  a_{i-1}..a_0 - 1}{2^i}⌋ + 1\right).
\end{align}
$

Now, noting:

- $
\begin{align}
⌊\frac{N -  a_{i-1}..a_0 - 1}{2^i}⌋
    =
        ⌊\frac{N}{2^i} - \frac{a_{i-1}..a_0 + 1}{2^i}⌋
    =
        \left\lfloor
            ⌊\frac{N}{2^i}⌋ + \{\frac{N}{2^i}\} - \frac{a_{i-1}..a_0 + 1}{2^i}
        \right\rfloor
    =
        ⌊\frac{N}{2^i}⌋ - (0 \;\text{or}\; 1),
\end{align}
$

- $
\begin{align}
Nx = N \cdot 0.a_0...a_{k-1} + \frac{N}{2^k} \cdot 0.a_k... = (\sum_{i} a_i \frac{N}{2^i}) + \eta
\end{align}
$
    with $\eta \lt 1$,

we got:

$
\begin{align}
B_N(x) - N x
    &=
        - \eta
        + \sum_{i = 0}^{k-1}
            a_i \cdot
                \left(
                    ⌊\frac{N -  a_{i-1}..a_0 - 1}{2^i}⌋ + 1
                    - \frac{N}{2^i}
                \right)
\\
    &=
        - \eta
        + \sum_{i = 0}^{k-1}
            a_i \cdot
                \left(
                    ⌊\frac{N}{2^i}⌋ - (0 \;\text{or}\; 1) + 1
                    -
                        \left(
                            ⌊\frac{N}{2^i}⌋ + \{ \frac{N}{2^i} \}
                        \right)
                \right)
\\
    &=
        - \eta
        + \sum_{i = 0}^{k-1}
            a_i \cdot
                \left(
                    + (1 \;\text{or}\; 0)
                    - \left\{ \frac{N}{2^i} \right\}
                \right).
\end{align}
$

Thus, 

$
\begin{align}
|B_N(x) - N x| \le 1 + \sum_{i = 0}^{k-1} 1 \cdot 1 \le k + 1 \le \log_2(N) + 2.
\end{align}
$

```python
#
# IEEE float bits manipulation
#
import numpy as np


def int_to_bits(x, n=32): # int -> str
    s = ''
    for i in range(n):
        if i > 0 and i % 4 == 0:
            s = ' ' + s
        b = (x >> i) & 1
        s = f"{b}" + s
    return s


def np_to_bits(x): # np -> str
    return int_to_bits(np.frombuffer(bytes(np.array([x])), np.int32)[0])


def bits_to_int(s): # str -> int
    s = s.replace(' ', '')    
    x = 0
    for i, c in enumerate(s[::-1]):
        if c == '1':
            x += (1 << i)
    return x


def reinterpret_cast(x, t1, t2):
    return np.frombuffer(bytes(np.array([x], dtype=t1)), dtype=t2)[0]


print(' 13 =', int_to_bits(13))
print('1.0 =', np_to_bits(np.float32(1)))

s = '0011 1111 0000 0000 0000 0000 0000 0000'
print(reinterpret_cast(bits_to_int(s), np.int32, np.float32), '=', s)
```

```python
#
# Fixed length bits reversal (actually this supports numpy array as well)
#

def bit_reversal8(x): # int -> int
    x = ((x & bits_to_int('0101 0101')) << 1) | ((x & bits_to_int('1010 1010')) >> 1)
    x = ((x & bits_to_int('0011 0011')) << 2) | ((x & bits_to_int('1100 1100')) >> 2)
    x = ((x & bits_to_int('0000 1111')) << 4) | ((x & bits_to_int('1111 0000')) >> 4)    
    return x

def bit_reversal32(x): # int -> int
    x = ((x & 0x55555555) <<  1) | ((x & 0xaaaaaaaa) >>  1)  # 0x5 = '0101', 0xa = '1010'
    x = ((x & 0x33333333) <<  2) | ((x & 0xcccccccc) >>  2)  # 0x3 = '0011', 0xc = '1100'
    x = ((x & 0x0f0f0f0f) <<  4) | ((x & 0xf0f0f0f0) >>  4)
    x = ((x & 0x00ff00ff) <<  8) | ((x & 0xff00ff00) >>  8)    
    x = ((x & 0x0000ffff) << 16) | ((x & 0xffff0000) >> 16)
    return x

print(int_to_bits(bit_reversal8(bits_to_int('0111 1101')), n=8))
print(int_to_bits(bit_reversal32(bits_to_int('0100 1000 0111 1101'))))
```

```python
#
# van der Corput sequence (base 2)
#

import matplotlib.pyplot as plt
import ipywidgets as ipyw


def corput2(n):
    a = np.uint32(np.arange(1, n + 1))
    b = np.uint32(np.log2(a)) + 1
    r = bit_reversal32(a)
    c = (r >> (32 - b)) / 2**b  # or simply "r / 2**32"
    return c


# This visualization actually corresponds to 2D Hammersley set
def _0(n=8, ax=False):
    a = np.arange(1, n + 1)
    c = corput2(n)
    if not ax:
        fig, ax = plt.subplots(figsize=(5, 5))
    ax.set(xlim=[1, n], ylim=[0, 1])
    ax.scatter(a, c, s=5)    

    
def _1(n=8, ax=False):
    rng = np.random.default_rng(0)
    x = rng.random((n, 2))

    if not ax:
        fig, ax = plt.subplots(figsize=(5, 5))
    ax.set(xlim=[0, 1], ylim=[0, 1])
    ax.scatter(*x.T, s=5)


def _2(n):
    fig, axes = plt.subplots(1, 2, figsize=(9, 4))
    _0(n, axes[0])
    _1(n, axes[1])
    

    
_2(n=1024)

# UI
# ipyw.interact(_0, n=(0, 1024, 1))
```

```python
#
# Visualize discrepancy of van der Corput sequence as cumulative function
#
# NOTE: "star" discrepancy is the maximal vertical distance of cumulative function and y = x diagonal line
#

def plot_cdf(c, ax=False):
    n = len(c)
    c = np.sort(c)
    a = (np.arange(n) + 0) / n
    b = (np.arange(n) + 1) / n
    disc = np.max(np.array([np.abs(c - a), np.abs(c - b)])) # "sup" is attained at discontinuity
    
    if not ax:
        fig, ax = plt.subplots(figsize=(5, 5))
    ax.set(xlim=[0, 1], ylim=[0, 1], title=f"discrepancy = {disc:.5f}")
    ax.grid()
    ax.scatter(c, a, s=10)
    ax.scatter(c, b, s=10)


def _0(n=7):
    c = corput2(n)
    d = np.random.default_rng(1).random(n)
    fig, axes = plt.subplots(1, 2, figsize=(11, 5))
    plot_cdf(c, ax=axes[0])
    plot_cdf(d, ax=axes[1])


_0(n=32)

# UI
# ipyw.interact(_0, n=(0, 1024, 8))
```

```python
#
# Visualize discrepancy wrt number of samples (compared to log(N) / N)
#

def compute_discrepancy(c):
    n = len(c)
    c = np.sort(c)
    a = (np.arange(n) + 0) / n
    b = (np.arange(n) + 1) / n
    disc = np.max(np.array([np.abs(c - a), np.abs(c - b)]))    
    return disc


def _(N=8):
    c = corput2(N)
    discs = np.empty(N)

    ns = np.arange(1, N + 1)
    for i in ns:
        discs[i - 1] = compute_discrepancy(c[:i])
        
    fig, ax = plt.subplots(figsize=(7, 5))
    ax.set(ylim=[0, 0.52])
    ax.grid()
    ax.plot(ns, discs)
    ns += 1
    ax.plot(ns, np.log(ns) / ns)


_(N=256)

# ipyw.interact(_, N=(1, 1024, 8))
```

```python
#
# digits reversal for base != 2
#

def int_to_digits(x, base): # int -> list[int]
    assert x >= 0
    assert base >= 2
    ls = []
    while x > 0:
        x, r = divmod(x, base)
        ls += [r]
    return ls[::-1]


def digits_reversal(x, base): # int -> int
    assert x >= 0
    assert base >= 2
    rev = 0
    while x > 0:
        x, r = divmod(x, base)
        rev = base * rev + r
    return rev


x = 1 * 4**3 + 0 * 4**2 + 2 * 4**1 + 3
int_to_digits(x, base=4), \
int_to_digits(digits_reversal(x, base=4), base=4)
```

```python
#
# Corput sequence with base != 2
#

def corput_single(x, base): # int -> float
    rev = 0
    pow_base = 1
    while x > 0:
        x, r = divmod(x, base)
        rev = base * rev + r
        pow_base *= base
    return rev / pow_base


def corput(n, base):  # int -> float32[..]
    c = [corput_single(x, base) for x in range(1, n + 1)]
    return np.float32(c)


def _0(n=8, base=3, ax=False):
    a = np.arange(1, n + 1)
    c = corput(n, base)
    if not ax:
        fig, ax = plt.subplots(figsize=(5, 5))
    ax.set(xlim=[1, n], ylim=[0, 1])
    ax.scatter(a, c, s=5)    


_0(n=1024)

# UI
# ipyw.interact(_0, n=(0, 1024, 1), base=(2, 17, 1))
```
