### General microfacet brdf formula

Writing

- $n \in H^2$ : normal
- $m \in H^2$ : microfacet normal
- $\omega \in H^2$ : out-going vector
- $D(m) \; [1/sr . m\text{-area} / n\text{-area}]$ : distribution of normals
- $dA_n$ : $n$-area of surface
- $dA_m$ : $m$-area of surface with microfacet normal $m$ within $dA_m$
- $dA_{\omega}$ : $\omega$-area from $dA_m$
- $L(\omega)$ : out-radiance from $dA_n$ surface
- $L(\omega \mid m)$ : out-radiance from $dA_m$ surface
- $\Phi$ : out-flux from $dA_n$ surface
- $\Phi_m$ : out-flux from $dA_m$ surface

then, writing the output-power for the surface $dA_n$ as a whole and as pieces, we have:

$
\Phi \, dA_n = (\text{power of $dA_n$}) = \sum_{m} \Phi_m \, dA_m.
$

Here by definition we have:

$
\begin{align}
\Phi \, dA_n
    &=
        L(\omega) dA_\omega d\omega
\\
\Phi_m \, dA_m
    &=
        L(\omega \mid m) dA_\omega d\omega
\\
dA_m
    &= dm \, dA_n D(m),
\end{align}
$

thus:

$
\begin{align}
L(\omega) dA_\omega d\omega
    =
        \Phi \, dA_n
    =
        \sum_m \Phi_m \, dA_m
    =
        \sum_m dA_m \frac{\Phi_m \, dA_m}{dA_m}
    =
        \sum_m dm \, dA_n D(m) L(\omega \mid m) \frac{dA_\omega}{dA_m} d\omega.
\end{align}
$

By divinding $dA_n d\omega$, we got:

$
\begin{align}
L(\omega) \frac{dA_\omega}{dA_n}
    =
        \sum_m dm \, D(m) L(\omega \mid m) \frac{dA_\omega}{dA_m}.
\end{align}
$

Finally taking limit:

$
\begin{align}
L(\omega) \, (n \cdot \omega)
    =
        \int_{dm} D(m) L(\omega \mid m) \, (m \cdot \omega).
\end{align}
$

Now, supposing micro-surface's brdf is given by $f$ and shadowing factor $G$ as in:

$
L(\omega_o | m) =
    \int_{d\omega_i} f(m, \omega_o, \omega_i) G(m, \omega_o, \omega_i) L_i(\omega_i) \, (m \cdot \omega_i)
$,

then, our formula becomes:

$
\begin{align}
L(\omega_o) \, (n \cdot \omega_o)
    &= 
        \int_{dm} D(m) L(\omega_o \mid m) \, (m \cdot \omega_o)
\\
    &= 
        \int_{dm} D(m) \, (m \cdot \omega_o)
            \int_{d\omega_i} f(m, \omega_o, \omega_i) G(m, \omega_o, \omega_i) L_i(\omega_i) \, (m \cdot \omega_i)
\\
    &= 
        \int_{d\omega_i} 
            \bigg(
                \int_{dm}
                    D(m) G(m, \omega_o, \omega_i) f(m, \omega_o, \omega_i)
                        \, (m \cdot \omega) \, (m \cdot \omega_i)
            \bigg)
                L_i(\omega_i)
\\
L(\omega_o)
    &= 
        \int_{d\omega_i} 
            \bigg(
                \frac{
                    \int_{dm}
                        D(m) G(m, \omega_o, \omega_i) f(m, \omega_o, \omega_i)
                        \, (m \cdot \omega) \, (m \cdot \omega_i)                        
                     }
                     {(n \cdot \omega_o) \, (n \cdot \omega_i)}
            \bigg)
                L_i(\omega_i)
                 \, (n \cdot \omega_i).
\end{align}
$

### Microfacet mirror brdf

We define mirro brdf by:

$
\begin{align}
f(m, \omega_o, \omega_i)
    =
        \frac{\delta(\omega_i - \text{refl}(\omega_o \mid m))}
             {(m \cdot \omega_i)},
\end{align}
$

where

$
\text{refl}(\omega_o \mid m)
    = (m \cdot \omega_o) m - (\omega_o - (m \cdot \omega_o) m)
    = 2 (m \cdot \omega_o) m - \omega_o.
$

Noting that 

$
\begin{align}
\int_{x} g(x) \delta(h(x))
    &= \int_{y} g(h^{-1}(y)) \delta(y) |\partial_y h^{-1}(y)|
        \quad \big(∵ \text{change of var.} \; x = h^{-1}(y) \big)    
\\
    &=
        g(\hat{x}) |\partial_x h(\hat{x})|^{-1}
        \quad \big( \text{where $\hat{x}$ s.t.}  \; h(\hat{x}) = 0\big).
\end{align}
$

we can evaluate the integral:

$
\begin{align}
\int_{dm}
    D(m) &G(m, \omega_o, \omega_i) f(m, \omega_o, \omega_i)
        \, (m \cdot \omega) \, (m \cdot \omega_i)
\\
    &=
         D(m) G(m, \omega_o, \omega_i)
             \, (m \cdot \omega)
             \, |\partial_m \text{refl}(\omega_o \mid m)|^{-1}.
\end{align}
$

where $m$ s.t. $\text{refl}(\omega_o \mid m) = \omega_i$ (i.e. $m = \text{half}(\omega_o, \omega_i)$.

To evaluate Jacobian, we note:

$
\begin{align}
\partial_m \text{refl}(\omega_o \mid m) \; \in R^{3 \times 3} \;
    &=
        \partial_m 2 (m \cdot \omega_o) m - \omega_o
     =
        2 ( (m \cdot \omega_o) I + m \omega_o^T)
\end{align}
\newcommand{\veca}[0]{\partial_r}
\newcommand{\vecb}[0]{\partial_\theta}
\newcommand{\vecc}[0]{\partial_\phi}
$

and writing orthognal vectors $\partial_r$, $\partial_\theta$, $\partial_\phi$ with $m = \partial_r$,
and $a = (\veca \cdot \omega_o), b = (\veca \cdot \omega_o), c = (\veca \cdot \omega_o)$,
(note that $|\partial_\phi| = \sin(\theta)$)

then we have:

$
\begin{align}
\partial_m \text{refl}(\omega_o \mid m) \, \partial_{\theta, \phi} m
        \; \in R^{3 \times 2} \;
    &=
        2 ( (\veca \cdot \omega_o) I + \veca \omega_o^T)
            \begin{pmatrix}
                \vecb, \vecc
            \end{pmatrix}
\\
    &=
        2 
            \big(
                a
                    \begin{pmatrix}
                        \vecb, \vecc
                    \end{pmatrix}        
                + \veca
                    \begin{pmatrix}
                        b & c
                    \end{pmatrix}                  
            \big)
\end{align}
$

$
\begin{align}
\big(
    \partial_m \text{refl}(\omega_o \mid m) \, &\partial_{\theta, \phi} m
\big)^T
\partial_m \text{refl}(\omega_o \mid m) \, \partial_{\theta, \phi} m
        \; \in R^{2 \times 2} \;
\\
    &=
        4
            \big(
                a^2
                \begin{pmatrix}
                    |\vecb|^2 & 0  \\
                           0  & |\vecc|^2
                \end{pmatrix}                
                +
                |\veca|^2
                \begin{pmatrix}
                    b^2 & b c  \\
                    b c & c^2
                \end{pmatrix}                
            \big)
\\
    &=
        4
            \begin{pmatrix}
                a^2 + b^2 & b c \\
                b c       & |\vecc|^2 a^2 + c^2
            \end{pmatrix}                
\end{align}
$

Thus, 

$
\begin{align}
| \partial_m \text{refl}(\omega_o \mid m) \, \partial_{\theta, \phi} m |^2
    &=
        4^2 ((a^2 + b^2) (|\vecc|^2 a^2 + c^2) - b c b c)
\\
    &=
        4^2 (|\vecc|^2 a^4 + a^2 (|\vecc|^2 b^2 + c^2))
\\
    &=
        4^2 a^2 \sin^2(\theta).
\end{align}
$

So,

$
\begin{align}
| \partial_m \text{refl}(\omega_o \mid m) | 
    &=
        \frac{| \partial_m \text{refl}(\omega_o \mid m) \, \partial_{\theta, \phi} m|}
             {|\partial_{\theta, \phi} m|}
\\
    &= 4 a = 4 (m \cdot \omega_o).
\end{align}
$

Inserting into our original integral, we get:

$
\begin{align}
\int_{dm}
    D(m) &G(m, \omega_o, \omega_i) f(m, \omega_o, \omega_i)
        \, (m \cdot \omega) \, (m \cdot \omega_i)
\\
    &=
         D(m) G(m, \omega_o, \omega_i)
             \, (m \cdot \omega)
             \, |\partial_m \text{refl}(\omega_o \mid m)|^{-1}
            \quad \big( m = \text{half}(\omega_o, \omega_i) \big)
\\
    &=
         \frac{D(m) G(m, \omega_o, \omega_i)}{4},
\end{align}
$

Or, as "total" brdf:

$
\begin{align}
L(\omega_o)
    &= 
        \int_{d\omega_i} 
            \bigg(
                \frac{
                    \int_{dm}
                        D(m) G(m, \omega_o, \omega_i) f(m, \omega_o, \omega_i)
                        \, (m \cdot \omega) \, (m \cdot \omega_i)                        
                     }
                     {(n \cdot \omega_o) \, (n \cdot \omega_i)}
            \bigg)
                L_i(\omega_i)
                 \, (n \cdot \omega_i)
\\
    &= 
        \int_{d\omega_i} 
            \bigg(
                \frac{D(m) G(m, \omega_o, \omega_i)}
                     {4 (n \cdot \omega_o) \, (n \cdot \omega_i)}
            \bigg)
                L_i(\omega_i)
                 \, (n \cdot \omega_i).
            \quad \big( m = \text{half}(\omega_o, \omega_i) \big)                 
\end{align}
$

### Derive $G$ from $P_2(q)$ and $\rho(h)$

$
\newcommand{\Pr}[0]{\text{Pr}}
$

Given

- $P_{22}(p, q) = P_2(p) P_2(q)$ : (isotropic) slope distribution,
- $\rho(h)$ : height distribution wrt. 2d plane,
- $m \in H^2$ : microfacet normal,
- $\omega \in H^2$ and $\mu = \cot(\text{acos}(n \cdot \omega))$,

Define

- $G(l, h, \mu, m)$ : ???,
- $G(h, \mu, m) = \int_{l \ge 0} G(l, h, \mu, m)$,
- $G(\mu, m) = \int_{h} \rho(h) G(h, \mu, m)$,

We model this phenomena by:

$
\begin{align}
G(l + \delta) - G(l)
    &=
        - \Pr[\text{occluded at $l + \delta$} \mid \text{not occluded at $l$}]
\\
    &=
        - \frac{\Pr[\text{not occluded at $l$} \wedge \text{occluded at $l + \delta$}]}
               {\Pr[\text{not occluded at $l$}]}
\end{align}
$

thus,

$
\begin{align}
\partial_l G(l)
    &= - 
        \frac{1}{\Pr[\text{not occluded at $l$}]}
        \lim_{\delta \downarrow 0}
            \frac{\Pr[\text{not occluded at $l$} \wedge \text{occluded at $l + \delta$}]}
                 {\delta}
\\
    &\equiv g(l)
\end{align}
$

and from $G(0) = \chi^+ (m \cdot \omega) \equiv G_0$ we have:

$G(l, h, \mu) = G_0 \exp\big(- \int_{l' \ge 0} g(l' | h, \mu)\big)$.

Analyzing those probabilities to take limit inside of $g(l)$:

$
\begin{align}
\Pr[\text{not occluded at $l$}]
    &=
        \int_{h' \le h + \mu l} \rho(h)
\\
    &\equiv F_\rho (h + \mu l)  \quad \big( \text{CDF of $\rho$} \big)
\\
\Pr[\text{not occluded at $l$} &\wedge \text{occluded at $l + \delta$}]
\\
    &\simeq
        \int_{p \ge \mu} P_2(p)
        \int_{\substack{h' \le h + \mu l \\ (h + \mu l - h') \le (p - \mu) \delta}} \rho(h')
\\
    &=
        \int_{p \ge \mu} P_2(p)
        \int_{h + \mu l - (p - \mu) \delta \le h' \le h + \mu l} \rho(h')        
\\
    &=
        \int_{p \ge \mu} P_2(p)
        \int_{h' \in [- (p - \mu) \delta, 0 ]} \rho(h' + h + \mu l)
\end{align}
$

thus,

$
\begin{align}
\lim_{\delta \downarrow 0}
    &\frac{\Pr[\text{not occluded at $l$} \wedge \text{occluded at $l + \delta$}]}
         {\delta}
\\
    &=
        \int_{p \ge \mu} P_2(p)
        \lim_{\delta \downarrow 0}
            \frac{\int_{h' \in [- (p - \mu) \delta, 0 ]} \rho(h' + h + \mu l)}
                 {\delta}
\\
    &=
        \int_{p \ge \mu} P_2(p) (p - \mu) \rho(h + \mu l)
\\
    &=
        \mu \rho(h + \mu l) \Lambda(\mu)      
\end{align}
$

where

- $\Lambda(\mu) = \frac{1}{\mu} \int_{p \ge \mu} P_2(p) (p - \mu)$.

So, we obtain:

$g(l) = \frac{\mu \rho(h + \mu l)}{F_\rho (h + \mu l)} \Lambda(\mu)$.

By noting:

- $\partial_l F_\rho (h + \mu l) = \mu \rho(h + \mu l)$,
- $\frac{\mu \rho(h + \mu l)}{F_\rho (h + \mu l)} = \partial_l \log F_\rho (h + \mu l)$,

we have,

$
\begin{align}
\int_{l \ge 0} g(l)
    &=
        \int_{l \ge 0} \frac{\mu \rho(h + \mu l)}{F_\rho (h + \mu l)} \Lambda(\mu)
\\
    &=
        \Lambda(\mu) \log \frac{F_\rho (\infty)}{F_\rho (h)}
\\
    &=
        - \Lambda(\mu) \log F_\rho (h)
\\
G(h, \mu, m) 
    &=
        G_0 \exp(- \int_{l \ge 0} g(l))
\\
    &=
        G_0 \exp \big( \Lambda(\mu) \log F_\rho (h) \big)
\\
    &=
        G_0 \big( F_\rho (h) \big)^{\Lambda(\mu)}
\\
G(\mu, m)
    &=
        G_0 \int_h \rho(h) \big( F_\rho (h) \big)^{\Lambda(\mu)},
\\
    &=
        G_0  \left. \frac{1}{1 + \Lambda(\mu)} F_\rho (h) \right|^{\infty}_{-\infty}
\\
    &=
        G_0  \frac{1}{1 + \Lambda(\mu)}
\\
\end{align}
$

### $P_{22}(p, q)$ to $D(m)$

By definition,

- $D(m) (n \cdot m)$ : distribution of micro noraml on $H^2$.

We also define:

- $\rho(\theta, \phi)$ :
    distribution parametrized by spherical coordinates
    $\theta \in [0, \frac{\pi}{2}]$ and $\phi \in [0, 2 \pi]$,
- $P_{22}(p, q)$ : distribution parametrized as slope $\in R^2$.

So, we just need to follow Jacobians.

First, we know:

- $|\partial_{\theta, \phi} m| = \sin(\theta)$

Second, we know the relation betw $\theta, \phi$ and slopes $p, q$ as:

$
\begin{align}
    \begin{pmatrix}
        \sin(\theta) \cos(\phi) \\
        \sin(\theta) \sin(\phi) \\
        \cos(\theta)
    \end{pmatrix}
    &\parallel
        \begin{pmatrix}
            1 \\ 0 \\ p
        \end{pmatrix}
        \times
        \begin{pmatrix}
            0 \\ 1 \\ q
        \end{pmatrix}    
    =
        \begin{pmatrix}
            -p \\ -q \\ 1
        \end{pmatrix}
\\
    &=
        \frac{1}{\sqrt{p^2 + q^2 + 1}}
            \begin{pmatrix}
                -p \\ -q \\ 1
            \end{pmatrix}
\\            
\cos(\theta) &= \frac{1}{\sqrt{p^2 + q^2 + 1}}
\\
\sin^2(\theta) &= \frac{p^2 + q^2}{p^2 + q^2 + 1}
\\
\tan^2(\theta) &= p^2 + q^2
\end{align}
$

So, conversely:

$
\begin{align}
    \begin{pmatrix}
        p \\ q
    \end{pmatrix}
    &=
        \sqrt{p^2 + q^2 + 1}
        \begin{pmatrix}
            \sin(\theta) \cos(\phi) \\
            \sin(\theta) \sin(\phi) \\
        \end{pmatrix}
    =
        \frac{1}{\cos(\theta)}
        \begin{pmatrix}
            \sin(\theta) \cos(\phi) \\
            \sin(\theta) \sin(\phi) \\
        \end{pmatrix}
    =
        \tan(\theta)
        \begin{pmatrix}
            \cos(\phi) \\ \sin(\phi)
        \end{pmatrix}
\end{align}
$


Thus,

$
\begin{align}
\partial_{\theta, \phi} \begin{pmatrix} p \\ q \end{pmatrix}
    &=
        \begin{pmatrix}
            \frac{1}{\cos^2(\theta)} \cos(\phi) & \tan(\theta) (-\sin(\phi)) \\
            \frac{1}{\cos^2(\theta)} \sin(\phi) & \tan(\theta) (\cos(\phi))
        \end{pmatrix}
\\
\left| \left.
    \partial_{\theta, \phi} \begin{pmatrix} p \\ q \end{pmatrix}        
\right. \right|
    &= \frac{1}{\cos^2(\theta)} \tan(\theta)
     = \frac{\sin(\theta)}{\cos^3(\theta)}
\end{align}
$

Now, combining them all, we obtain:

$
\begin{align}
D(m) (n \cdot m)
    &=
        P_{22}(p, q) 
        \left| \left.
            \partial_{\theta, \phi} \begin{pmatrix} p \\ q \end{pmatrix}        
        \right. \right|
        \left| \left.
            \partial_{\theta, \phi} m
        \right. \right|^{-1}     
     =
        P_{22}(p, q) \frac{1}{\cos^3(\theta)}
\\
D(m)
    &=
        P_{22}(p, q) \frac{1}{\cos^4(\theta)}.
\end{align}
$

### GGX

We notice it's non separable distribution (note that Beckman $P_{22}$ was separable.):

$

Now, before getting into $\Lambda$, we first note that "variance transformation rule" of $\Lambda$.

$
\begin{align}
\Lambda(\mu \mid \alpha)
    &\equiv

            \int_{t \ge \frac{\mu}{\alpha}}
                \alpha \frac{1}{\alpha} P_2(t) (t - \frac{\mu}{\alpha})
\\
    &= \Lambda(\frac{\mu}{\alpha}).
\end{align}
$

\\
    &=
        \int_{\theta \in [\atan(\mu), \frac{\pi}{2}]}
            (\tan(\theta) - \mu) \cos(\theta)
\\
    &=
        \int_{\theta \in [\atan(\mu), \frac{\pi}{2}]}
            \sin(\theta) - \mu \cos(\theta)
\\
    &=
        \left.
            - \cos(\theta) - \mu \sin(\theta)                
        \right|_{\atan(\mu)}^{\frac{\pi}{2}}    
\\
    &=
        - \mu + \cos(\theta_0) + \mu \sin(\theta_0)
        \quad \big( \text{where} \; \theta_0 = \atan(\mu) \big)
\\
    &=
        - \mu + \cos(\theta_0) (1 + \mu \tan(\theta_0))
\\
    &=
        - \mu + \frac{1}{\sqrt{1 + \mu^2}}(1 + \mu^2)
\\
    &=
        - \mu + \sqrt{1 + \mu^2},
\end{align}
$

we get:

$
\begin{align}
\Lambda(\mu)
    &=
        \frac{1}{\mu} \int_{p \ge \mu} P_2(p) (p - \mu)    
\\
    &=
        \frac{1}{2 \mu}
            \int_{p \ge \mu}
                \big(
                    \frac{1}{1 + p^2}
                \big)^{\frac{3}{2}}
                (p - \mu)
\\
    &=
        \frac{1}{2} \frac{- \mu + \sqrt{1 + \mu^2}}{\mu}
\\
    &=
        \frac{- 1 + \sqrt{\frac{1}{\mu^2} + 1}}{2} 
\end{align}
$

### CDF of $D(m) (n \cdot m)$ (GGX)

We write distribution parametrized by $\theta$ and $\phi$:

$
\begin{align}
\rho(\theta, \phi)
    &=
        \sin(\theta) D(m) (n \cdot m)
\\
    &=
        \sin(\theta) \cos(\theta)
            \frac{\alpha^2}{\pi}
            \frac{1}{\big((\alpha^2 - 1) \cos^2(\theta) + 1\big)^2}.
\end{align}
$

So, it's CDF along $\theta \in [0, \frac{\pi}{2}]$ is:

$
\begin{align}
F(t)
    &\equiv
        \int_{\theta \le t} \int_{\phi} \rho(\theta, \phi)
\\
    &= 2 \pi
        \int_{\theta \le t} 
            \sin(\theta) \cos(\theta)
                \frac{\alpha^2}{\pi}
                \frac{1}{\big((\alpha^2 - 1) \cos^2(\theta) + 1\big)^2}
\\
    &= 2 \alpha^2
        \int_{\theta \le t} 
            \sin(\theta) \cos(\theta)
                \frac{1}{\big((\alpha^2 - 1) \cos^2(\theta) + 1\big)^2}
\\
    &= 2 \alpha^2
        \left.
            \frac{1}{2 (\alpha^2 - 1)}
            \big(
                (\alpha^2 - 1) \cos^2(\theta) + 1
            \big)^{-1}
        \right|_{0}^{t}
\\
    &= \frac{\alpha^2}{\alpha^2 - 1}
        \big(
            \frac{1}{(\alpha^2 - 1) \cos^2(t) + 1} - \frac{1}{\alpha^2}
        \big)
\\
    &= \frac{1}{\alpha^2 - 1}
        \big(
            \frac{\alpha^2}{(\alpha^2 - 1) \cos^2(t) + 1} - 1
        \big)        
\\
    &= \frac{1}{\alpha^2 - 1}
        \big(
            \frac{\alpha^2 - (\alpha^2 - 1) \cos^2(t) - 1}{(\alpha^2 - 1) \cos^2(t) + 1}
        \big)            
\\
    &= 
        \frac{1 - \cos^2(t)}{(\alpha^2 - 1) \cos^2(t) + 1}
     = 
        \frac{\sin^2(t)}{\alpha^2 \cos^2(t) + 1 - \cos^2(t)}    
     = 
        \frac{\sin^2(t)}{\alpha^2 \cos^2(t) + \sin^2(t)}
     = 
        \frac{1}{\frac{\alpha^2}{\tan^{2}(t)} + 1}.
\end{align}
$

Also, its inverse can be seen:

$
\begin{align}
\frac{1}{F(t)} &= \frac{\alpha^2}{\tan^{2}(t)} + 1
\\
\frac{1}{\frac{1}{F(t)} - 1} &= \frac{\tan^{2}(t)}{\alpha^2}
\\
\tan^{2}(t) &= \alpha^2 \frac{1}{\frac{1}{F(t)} - 1}
\\
\tan(t) &= \alpha \sqrt{\frac{F(t)}{1- F(t)}}.
\\
\end{align}
$
