## Euler's equation

### Prop.

Given $A(t) \in \tt{SO}(3)$, for some $\omega(t) \in R^3$, we have:

$~~
\bba
\dot{A}   &= A C_{\omega(t)} \in R^{3 \times 3},  \\
\eea
$

where

$~~
C_v = \ll(\bbm
       0 & -v_z &  v_y \\
     v_z &    0 & -v_x \\
    -v_y &  v_x & 0
\eem\rr)
~~ (\tt{i.e.} C_v x = v \times x).
$


***Proof***

Since $A^{T} A = I$, we have:

$~~
\bba
A^{T} A = I
    &\implies  \pp_t \ll( A^{T} A \rr) = 0
    \iff  \dot{A}^T A + A^T \dot{A} = 0
    \iff  (A^T \dot{A})^T  + A^T \dot{A} = 0.
\eea
$

Thus, $A^T \dot{A} = A^{-1} \dot{A}$ : anti-symmetric.

### Prop.

Given $A \in \tt{SO}(3)$ and $u, v \in R^3$, we have:

$~~A (u \times v) = A u \times A v$.

***Proof***

Using generalized Kronecker delta notation, we see:

$~~
\bba
A u \times A v = (\Fb{a_i} u_i) \times (\Fb{a_j} u_j) = \delta_{i,j,k}^{1, 2, 3} \Fb{a_k} u_i u_j = A (u \times v).
\eea
$

### Def.

Given

- $V \subset R^3$,
- $\rho(x) : R^3 \to R$,
- $G_t(x) : R \times R^3 \to R^3$ (below occasionally we omit argument $t$),

we define:

$~~
L[G_t] \equiv \int_{x \in V} \rho(x) G_t(x) \times \partial_t G_t(x).
$

### Prop.

Given

- $A_t \in \tt{SO}(3)$ with corresponding $\omega_t$ from previous proposition

then,

$~~
L[A_t] = A_t I \omega_t
$

where

$~~
\bba
I \equiv \int_{x \in V} \rho(x) \ll( |x|^2 \tt{id} - x x^T \rr)
  = \int_{x \in V} \rho(x) \ll(\bbm 
        y^2 + z^2 & xy        & zx \\
        xy        & z^2 + x^2 & xy \\
        zx        & xy        & x^2 + y^2
    \eem\rr)
  ~ \in R^{3 \times 3}.
\eea
$

***Proof***

First we note:

$~~
\bba
A(x) \times \pp_t A(x)
    = Ax \times (A \omega \times A x)
    = A (x \times (\omega \times x))
    = A (|x|^2 id - x x^T) \omega.
\eea
$

Thus, we have:

$~~
\bba
L[A_t]
    &= \int_{x \in V} \rho(x) Ax \times \pp_t Ax \\
    &= \int_{x \in V} \rho(x) A (|x|^2 id - x x^T) \omega \\
    &= A \ll( \int_{x \in V} \rho(x) \ll( |x|^2 id - x x^T \rr) \rr) \omega.
\eea
$

### Def.

Define $\tau$ by:

$~~
\bba
\tau[G_t] \equiv \int_{x \in V} G_t(x) \times \rho(x) \pp_t^2 G_t(x).
\eea
$


### Prop.

$~~
\tau[G_t] = \pp_t L[G_t].
$

***Proof***

Noting that:

$~~
\pp_t (G(x) \times \pp_t G(x))
    = \pp_t G(x) \times \pp_t G(x) + G(x) \times \pp_t^2 G(x)
    = G(x) \times \pp_t^2 G(x),
$

we have:

$~~
\bba
\pp_t L[G_t]
    &= \int_{x \in V} \rho(x) \pp_t \ll( G(x) \times \pp_t G(x) \rr)
     = \int_{x \in V} \rho(x) G(x) \times \pp_t^2 G(x).
\eea
$

### Prop.

Given

- $G_t(x) \in R^3$,
- $M = \int_{x \in V} \rho(x)$,
- $b_t = \ll( \int_{x \in V} \rho(x) G_t(x) \rr) / M$,
- $A_t(x) = G_t(x) - b_t$ with $A_t \in \tt{SO}(3)$,

we have:

$~~
\bba
\tau[G_t]
    &= b_t \times \pp_t^2 \ll( \int_{x} \rho(x) G_t(x) \rr)
        + \int_{x} \rho(x) A(x) \times \pp_t^2 G(x) \\
    &= b_t \times M \ddot{b_t} 
        + \int_{x} \rho(x) A(x) \times \pp_t^2 G(x).
\eea
$

Also,

$~~
\bba
\int_{x} \rho(x) A(x)
    = \int_{x} \rho(x) (G(x) - b)
    = \ll( \int_{x} \rho(x) G(x) \rr) - M b = 0,
\eea
$

thus,

$~~
\bba
\int_{x} \rho(x) A(x) \times \pp_t^2 G(x)
    &= \ll( \int_{x} \rho(x) A(x) \rr) \times \ddot{b} + \int_{x} \rho(x) A(x) \times \pp_t^2 A(x) \\
    &= \int_{x} \rho(x) A(x) \times \pp_t^2 A(x) = \pp_t L[A_t] = \pp_t \ll( A_t I \omega_t \rr).
\eea
$

In summary, we got Euler's equation:

$~~
\bba
\tau[G_t] - b_t \times M \ddot{b_t}
    &= \int_{x} \rho(x) A(x) \times \pp_t^2 G(x) \\
    &= \pp_t \ll( A_t I \omega_t \rr).
\eea
$

## Energy

### Def.

Define $E$ by:

$~~
E[G] \equiv \int_x \frac{1}{2} \rho(x) \ll| \pp_t G_t(x) \rr|^2.
$


### Prop.

Noting that:

$~~
\bba
\ll| \dot{A_t(x)} \rr|^2
    &= \ll| A (\omega \times x) \rr|^2 \\
    &= \ll| \omega \times x \rr|^2  ~~ (\tt{since $A$ : orthogonal}) \\
    &= \ll< \omega \times x, \omega \times x \rr> \\
    &= \ll< \omega,  x \times (\omega \times x) \rr> \\
    &= \ll< \omega,  \ll( |x|^2 \tt{id} - x x^T \rr) \omega \rr>,
\eea
$

and,

$~~
\int_x \rho(x) \dot{A_t(x)} = \pp_t \ll( \int_x \rho(x) (G_t(x) - b_t) \rr) = 0,
$

we have:

$~~
\bba
E[G] &=
    \frac{1}{2} M \dot{b_t}^2
        + \frac{1}{2} \int_x \rho(x) \dot{A_t(x)}^2
        + \int_x \rho(x) \dot{A_t(x)} \cdot \dot{b_t} \\
    &=
        \frac{1}{2} M \dot{b_t}^2
        + \frac{1}{2} \omega^T \ll( \int_x \rho(x) |x|^2 \tt{id} - x x^T \rr) \omega  \\
    &=
        \frac{1}{2} M \dot{b_t}^2
        + \frac{1}{2} \omega^T I \omega.
\eea
$

```python
#
# Torque free solution (plot trajectory of angular momentum)
#
# Dw = - I^-1 (w x Iw)
# DA = A Cw  (Cw is s.t. Cw v = cross(w, x))
#

import numpy as np; Np = np.array
import matplotlib.pyplot as plt
import mpl_toolkits.mplot3d

#
# Utilities
#

normalize = lambda x: x / np.linalg.norm(x)


# Explicit Runge-Kutta 4th order
def erk4(f, t, x, dt):
    k1 = f(t           , x                )
    k2 = f(t + 1/2 * dt, x + 1/2 * dt * k1)
    k3 = f(t + 1/2 * dt, x + 1/2 * dt * k2)
    k4 = f(t + 1   * dt, x + 1   * dt * k3)
    return x + dt * (1/6 * k1 + 2/6 * k2 + 2/6 * k3 + 1/6 * k4)


def solve(f, x, ls_t): # -> float[N, ...]
    N = len(ls_t)
    ls_dt = ls_t[1:] - ls_t[:-1]
    ls_x = [x]
    for i in range(N - 1):
        t = ls_t[i]
        dt = ls_t[i + 1] - t
        x = erk4(f, t, x, dt)
        ls_x += [x]
    return Np(ls_x)


#
# Numerical integration
#

# (diagonal) moments of inertia  (Ix > Iy > Iz)
I = Np([1, 2, 3]) 


def _0(w, ax=False):
    N = 200
    dt = 0.05
    f = lambda t, w: - (1 / I) * (np.cross(w, I * w))

    ls_t = np.arange(N) * dt
    ls_w = solve(f, w, ls_t)
    ls_m = ls_w * I  # angular moment 


    #
    # Visualization
    #

    x, y, z = ls_m.T

    if ax is False:
        fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(projection='3d'))

    elev=30
    azim=-50
    ax.view_init(elev, azim)
    ax.set(xlabel='x', ylabel='y', zlabel='z')
    ax.scatter(x, y, z, c=ls_t, s=10, cmap='hsv')

# Three initial values
# wx, wz: stable
# wy: unstable (cf. "Tennis racket theorem")
wx = normalize(Np([1.0, 0.2, 0.0])) * 2 * np.pi / 4
wy = normalize(Np([0.0, 1.0, 0.2])) * 2 * np.pi / 4
wz = normalize(Np([0.2, 0.0, 1.0])) * 2 * np.pi / 4

fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(projection='3d'))
lim = [-4, 4]
ax.set(xlim=lim, ylim=lim, zlim=lim, xlabel='x', ylabel='y', zlabel='z')
_0(wx, ax=ax)
_0(wy, ax=ax)
_0(wz, ax=ax)
```
