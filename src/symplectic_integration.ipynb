## Symplectic map and generating function

(note that we order coordinates in $R^{2n}$ by $(p_1, .., p_n, q_1, .., q_n)$.)

### Def. 1

$
\phi : R^{2n} \to R^{2n} : \tt{symplectic map}
\eqarrow
(D\phi)^T J D\phi = J
$

where 

- $D\phi$ : Jacobian of $\phi$,
- $J = \Mmp 0 & I \\ -I & 0 \mmp$.


### Prop. 2

$
\phi : \tt{symplectic map}
\eqarrow
\phi^{*} (dp_i \wedge dq_i) = dp_i \wedge dq_i
$

(note that "$dp_i \wedge dq_i$" denotes $\sum_i dp_i \wedge dq_i$)

***Proof***

We obtain result from:

$
\Aa
(dp_i \wedge dq_i) (u, v)
    = \sum_i u_{p_i} v_{q_i} - u_{q_i} v_{p_i}
    =  u^T J v,
\aa
$

and,

$
\Aa
(\phi^{*} (dp_i \wedge dq_i)) (u, v)
    = (dp_i \wedge dq_i) (D\phi(u), D\phi(v))
    = (D\phi(u))^T J D\phi(v).
\aa
$

### Prop. 3

$
\Aa
\phi : \tt{symplectic map}
&\eqarrow
    \phi^{*}(p_i dq_i) - p_i dq_i = dS \quad \tt{for some S} \\
&\eqarrow
    \phi^{*}(q_i dp_i) + p_i dq_i = dS \quad \tt{for some S} \\
&\eqarrow
    \phi^{*}(p_i dq_i) + q_i dp_i = dS \quad \tt{for some S}. \\
\aa
$


***Proof***

Noting:

$
\Aa
dp_i \wedge dq_i           &= d(p_i dq_i) = - d(q_i dp_i) \\
\phi^{*}(dp_i \wedge dq_i) &= d(\phi^{*}(p_i dq_i)) = - d(\phi^{*}(q_i dp_i)),
\aa
$

we have:

$
\Aa
\phi^{*}(dp_i \wedge dq_i) = dp_i \wedge dq_i
    &\eqarrow
        \phi^{*}(dp_i \wedge dq_i) - dp_i \wedge dq_i = 0 \\
    &\eqarrow
        \phi^{*}(p_i dq_i) - p_i dq_i : \tt{exact} \\
    &\eqarrow
        \phi^{*}(q_i dp_i) + p_i dq_i : \tt{exact} \\
    &\eqarrow
        \phi^{*}(p_i dq_i) + q_i dp_i : \tt{exact}.   
\aa
$

Then, "closed" $\eqarrow$ "exact" follows from Poincare's lemma.

### Prop. 4

We write $\phi(p, q) = (\hat{p}(p, q), \hat{q}(p, q))$.

We assume $\exists S : R^{2n} \to R$ s.t.

$
\Cc\Aa
            p &= \pp_{q} S(\hat{p}(p, q), q)  \\
\hat{q}(p, q) &= \pp_{\hat{p}} S(\hat{p}(p, q), q).
\aa\cc
$

Then, $\phi$ : symplectic map.


***Proof***

We define $\psi : R^{2n} \to R^{2n}$ s.t.

$
\psi(p, q) = (\hat{p}(p, q), q).
$

Then,

$
\Aa
d(\psi^*(S))
    &= \psi^*(dS) \\
    &= \psi^*(\pp_{p_i} S(p, q) dp_i + \pp_{q_i} S(p, q) dq_i) \\
    &= \pp_{\hat{p_i}} S(\hat{p}, q) d(\hat{p}_i) + \pp_{q_i} S(\hat{p}, q) dq_i \\
    &= \pp_{\hat{p_i}} S(\hat{p}, q) d(\hat{p}_i) + \pp_{q_i} S(\hat{p}, q) dq_i \\    
    &= \hat{q}_i d(\hat{p}_i) + p_i dq_i. \\
\aa
$

Thus, our result follows from "Prof. 3".

## Verlet integration


### Def. 1

Given a differential equation of the form:

$
\Aa
\pp_t x(t)
    = G(x(t)),
\aa
$

by separting components by $x = \Mmp p \\ q \mmp$, we also write:

$
\Aa
\pp_t \Mmp p \\ q \mmp
    = \Mmp G_p(p, q) \\ G_q(p, q) \mmp,
\aa
$

(note that $G_p$ denotes p-components of vector $G$ and it's not partial derivative,)

then, we define ***p-implicit method*** and ***q-implicit method*** as follows:

- p-implicit method:

$
\Cc\Aa
p_1 &= p_0 + h G_p(p_1, q_0) \\
q_1 &= q_0 + h G_q(p_1, q_0),
\aa\cc
$


- q-implicit method:

$
\Cc\Aa
p_1 &= p_0 + h G_p(p_0, q_1) \\
q_1 &= q_0 + h G_q(p_0, q_1),
\aa\cc
$

We also define ***pq-implicit method*** and ***qp-implicit method***
by alternating above methods in two steps.

- pq-implicit method

$
\Cc\Aa
p_1 &= p_0 + h G_p(p_1, q_0) \\
q_1 &= q_0 + h G_q(p_1, q_0) \\
p_2 &= p_1 + h G_p(p_1, q_2) \\
q_2 &= q_1 + h G_q(p_1, q_2),
\aa\cc
$

- qp-implicit method

$
\Cc\Aa
p_1 &= p_0 + h G_p(p_0, q_1) \\
q_1 &= q_0 + h G_q(p_0, q_1) \\
p_2 &= p_1 + h G_p(p_2, q_1) \\
q_2 &= q_1 + h G_q(p_2, q_1),
\aa\cc
$

### Prop. 2.

For p-implicit and q-implicit method, we have:

$
x(h) - x_1 = O(h^2).
$

(aka. convergence of 1st order.)

***Proof***

Here, we only show the proof for p-implicit method. 

Noting:

$
\Aa
G_p(p_1, q_0)
    &= G_p(p_0 + h G_p(p_1, q_0), q_0) \\
    &= G_p(p_0, q_0) + O((h G_p(p_1, q_0))) \\
    &= \dot{p}_0 + O(h),    
\aa
$

thus,

$
\Aa
p_1 &= p_0 + h G_p(p_1, q_0) = p_0 + h \dot{p}_0 + O(h^2)
\\
q_1 &= q_0 + h G_q(p_1, q_0) = q_0 + h \dot{q}_0 + O(h^2).
\aa
$

i.e. $x_1 = x_0 + h \dot{x}_0 + O(h^2)$.

On the other hand, by Taylor expansion, we have $x(h) = x(0) + h \dot{x}(0) + O(h^2)$, so our result follows.

### Prop. 3.

For pq-implicit method and qp-implicit method, we have:

$
x(2h) - x_2 = O(h^3).
$

(aka. convergence of 2nd order.)

***Proof***

Here, we only show the proof of qp-implicit method.

First, by Taylor expansion around $t = h$, we have:

$
x(2h) - x(0) = 2 h \dot{x}(h) + O(h^3).
$

On the other hand, we see

$
\Aa
p_2 - p_0 &= h ( G_p(p_0, q_1) + G_p(p_2, q_1) ) \\
q_2 - q_0 &= h ( G_q(p_0, q_1) + G_q(p_2, q_1) ),
\aa
$

and,

$
\Aa
G_p(p_0, q_1) + G_p(p_2, q_1)
    &= G_p(p_1 - h G_p(p_0, q_1), q_1) + G_p(p_1 + h G_p(p_2, q_1), q_1) \\
    &= 2 G_p(p_1, q_1) + h \pp_p G_p(p_1, q_1) (G_p(p_2, q_1) - G_p(p_0, q_1)) + O(h^2),
\aa
$

and,

$
\Aa
G_p(p_2, q_1) - G_p(p_0, q_1)
    &= G_p(p_1 + h G_p(p_2, q_1), q_1) - G_p(p_1 - h G_p(p_0, q_1), q_1) \\
    &= O(h),
\aa
$

thus,

$
G_p(p_0, q_1) + G_p(p_2, q_1)
    = 2 G_p(p_1, q_1) +  h \pp_p G_p(p_1, q_1) O(h) + O(h^2)
    = 2 G_p(p_1, q_1) + O(h^2),
$

and,

$
p_2 - p_0 = h ( G_p(p_0, q_1) + G_p(p_2, q_1) ) = 2 h G_p(p_1, q_1) + O(h^3).
$

therefore,

$
x_2 - x_0 = 2 h G(x_1) + O(h^3).
$

Finllay, noting:

$
G(x_1) = G(x_0 + O(h^2)) = G(x_0) + O(h^2),
$

we have:

$
x_2 - x_0 = 2 h G(x_1) + O(h^3) = 2 h G(x_0) + O(h^3) = 2 h \dot{x}_0 + O(h^3).
$

### Prop. 4.

Assuming $G(p, q)$ is given by:

- $G(p, q) = \Mmp G_p(p, q) \\ G_q(p, q) \mmp = \Mmp B(q) \\ p \mmp,$

then, the sequence of pq-implicit method $p_0, p_1, p_2, p_3, p_4$ and $q_0, q_1, q_2, q_3, q_4$ satisfies:

$
\Cc\Aa
q_2 &= q_0 + 2h p_0 + \frac{(2h)^2}{2} B(q_0)  \\
q_4 &= 2 q_2 - q_0 + (2h)^2 B(q_2)  \\
p_2 &= \frac{q_4 - q_0}{4 h}.
\aa\cc
$

(aka. Verlet scheme by replacing $\Delta_t = 2h$.)

***Proof***

For this special case, pq-implicit method gives:

$
\Cc\Aa
p_1 &= p_0 + h B(q_0)  \\
q_1 &= q_0 + h p_1     \\
p_2 &= p_1 + h B(q_2)  \\
q_2 &= q_1 + h p_1,
\aa\cc
$

Here, we have:

$
\Aa
p_1 &= p_2 - h B(q_2)  \\
p_3 &= p_2 + h B(q_2),
\aa
$

and:

$
\Aa
p_3 + p_1 &= 2 p_2 \\
p_3 - p_1 &= 2 h B(q_2).
\aa
$

Also, we have:

$
\Aa
q_2 - q_0 &= 2 h p_1 \\
q_4 - q_2 &= 2 h p_3.
\aa
$

Therfore:

$
\Aa
q_4 - 2 q_2 + q_0
    &= (q_4 - q_2) - (q_2 - q_0)
    =  2 h (p_3 - p_1)
    =  (2 h)^2 B(q_2),
\aa
$

and:

$
\Aa
q_4 - q_0
    &= (q_4 - q_2) + (q_2 - q_0)
    =  2 h (p_3 + p_1)
    =  4 h p_2,
\aa
$

and:

$
\Aa
q_2 - q_0
    &= 2 h p_1
    =  2 h p_0 + 4 h^2 B(q_0).
\aa
$
