## SOR method


### Def. 1

Given $A \in \ffbb{C}^{n \times n}$, we write $A = D + L + U$ where each D/L/U has diagonal/lower/upper parts of A.

Define ***w-SOR operator*** $x \mapsto x''$  by:

$~~
\bba
D x' = - L x'' - U x + b \\
x'' = (1 - w) x + w x',
\eea
$

i.e.

$~~
\bba
& D x'' = (1 - w) D x + w D x' = (1 - w) D x + w (- L x'' - U x + b) \\
& \begin{aligned}
    \iff
        \ll( \frac{1}{w} D + L \rr) x''
            &= (\frac{1 - w}{w} D - U) x + b
             = - \ll( \ll( 1 - \frac{1}{w} \rr) D + U \rr) x + b \\
            &= - \ll(A - \ll( \frac{1}{w} D + L \rr) \rr) x + b.
  \end{aligned}
\eea
$

When $w = 1$, it produces ***Gauss-Seidel operator***, which is defined by:

$~~ (D + L) x' + U x = b ~~$ i.e. $~~ x' = (D + L)^{-1}(- U x + b)$.

### Prop. 2.

(Cf. Stoer, Bulirsch Thm. 8.3.7.)

Given

- $A$ : positive definite,
- $w \in (0, 2)$,
- $B \equiv \frac{1}{w} D + L$,
- $H \equiv B^{-1}(B - A) = I - B^{-1} A$,

then $\rho(H) \lt 1$.

***Proof***

First we (magically) find that:

$~~
\bba
H &= I - B^{-1} A  \\
    &= (A^{-1} B - I) B^{-1} A  \\
    &= (A^{-1} B - I) (A^{-1} B)^{-1} \\
    &= (2 A^{-1} B - 2 I) (2 A^{-1} B)^{-1} \\    
    &= \ll( (2 A^{-1} B - I) - I \rr) ((2 A^{-1} B - I) + I)^{-1},
\eea
$

thus, writing $Q = 2 A^{-1} B - I$, we have $H = (Q - I) (Q + I)^{-1}$.

Now, given eigen value $\mu$ and eigen vector $x$ of $H$ as in $H x = \mu x$,
we define $y = (R + I)^{-1} x$ and we see:

$~~
\bba
H x = \mu x
    &\iff (Q - I) \ll( (Q + I)^{-1} x \rr) = \mu (Q + I) \ll( (Q + I)^{-1} x \rr) \\
    &\iff Qy - y = \mu Q y + \mu y \\
    &\iff Q y = \frac{1 + \mu}{1 - \mu} y.
\eea
$

Here, we observe:

$~~
\bba
\frac{1 + \mu}{1 - \mu}
    &= \frac{(1 + \mu)(1 - \mu^\dagger)}{|1 - \mu|^2}
     = \frac{1 + i \Im{\mu} - |\mu|^2}{|1 - \mu|^2} \\
\Re{\frac{1 + \mu}{1 - \mu}}
    &= \frac{1 - |\mu|^2}{|1 - \mu|^2}.
\eea
$

Thus, 

$~~
\Re{\frac{1 + \mu}{1 - \mu}} > 0  \iff  1 - |\mu|^2 > 0  \iff |\mu| < 1,
$

so, to our goal, it's suffices to show $\Re{\tt{eigen value of $Q$}} > 0$.

Proving this, we take $Q y = (2 A^{-1} B - I) y = \lambda y$, then we see:

$~~
\bba
&(2 A^{-1} B - I) y = \lambda y  \\
&\iff A^{-1} (2 B - A) y = \lambda y  \\
&\iff (2 B - A) y = \lambda A y \\
&\implies
    \begin{cases}
        \iprod{y}{(2 B - A) y}         &= \iprod{y}{\lambda A y} \\
        \iprod{y}{(2 B - A)^\dagger y} &= \iprod{y}{\lambda^\dagger A^\dagger y}
    \end{cases} \\
&\iff
    \begin{cases}
        \iprod{y}{(2 B - A) y}         &= \iprod{y}{\lambda A y} \\
        \iprod{y}{(2 B^\dagger - A) y} &= \iprod{y}{\lambda^\dagger A y}
    \end{cases} ~~~(\tt{since $A$ : Hermittian}) \\
&\implies
    \iprod{y}{\ll( 2 (B + B^\dagger) - 2 A \rr) y} = (\lambda + \lambda^\dagger) \iprod{y}{A y}
    ~~~(\tt{adding above two equations}) \\
&\implies
    \iprod{y}{\ll((B + B^\dagger) - A \rr) y} = \Re{\lambda}\iprod{y}{A y}
\eea
$

But here, $(B + B^\dagger) - A = \frac{2}{w} D + L + L^\dagger - A = \ll(\frac{2}{w} - 1\rr) D$ : positive definite.

Thus, we obtain $\Re{\lambda} > 0$.
